# **Exploración de los Datos**

**Número de Estudiantes según su programa académico:**

Se filtró la columna que muestra el programa académico del estudiante y se cuenta cuantas veces se repite en los datos, es decir, cuántos estudiantes están matriculados por programa. Se puede apreciar que la carrera de derecho es la que más estudiantes tiene matriculados y por el contrario, las carreras de comunicación social y periodismo, deporte y actividad fisica son algunas que cuentan solo con un estudiante que presentó la prueba saber pro.

```python
students_per_program = z['E_PRGM_ACADEMICO'].value_counts()
print("Número de estudiantes por programa académico:")
display(students_per_program)
```

**Analizando el desempeño de los estudiantes de Ingeniería Industrial en las pruebas Saber Pro:**

Primero filtramos la carrera de ingeniería industrial dentro de la columna de E_PRGM_ACADEMICO

```python
industrial = z[z["E_PRGM_ACADEMICO"] == "INGENIERIA INDUSTRIAL"]
display(industrial)
```

Contamos la cantidad de estudiantes de Ingeniería Industrial por cada desempeño dentro de la columna RENDIMIENTO_GLOBAL:

```python
rendimiento_industrial = industrial["RENDIMIENTO_GLOBAL"].value_counts().reset_index()
rendimiento_industrial.columns = ["Rendimiento", "Cantidad"]

display(rendimiento_industrial)
```

Se puede apreciar que en el programa de Ingeniería Industrial, con un total de 28.481 estudiantes (siguiendo el primer filtrado de numero de estudiantes por programa), refleja con 17.212 estudiantes lo cual es el 60,4% un rendimiento alto o medio-alto lo que significa un desempeño académico favorable en la mayoría. Sin embargo, el 39,6% restante (11.269 estudiantes) se ubica en niveles medio-bajo y bajo, lo que supone estos estudiantes no les fue bien en la prueba.
